```python
def wrangle(csv_path):
    df = pd.read_csv(csv_path,sep=";",low_memory=False)
    

    values = ['P2','P1', 'P1','temperature','humidity','height']
    df = df[df.value_type.isin(values) == True]
    

    df['value']=df.value.str.replace('[^\d.]', '',regex=True).astype(float)
    
    df['P2'] = np.where(df['value_type'] == "P2", df["value"], 'NaN')
    df['P1'] = np.where(df['value_type'] == "P1", df["value"], 'NaN')
    df['P0'] = np.where(df['value_type'] == "P0", df["value"], 'NaN')
    df['temperature'] = np.where(df['value_type'] == "temperature", df["value"], 'NaN')
    df['humidity'] = np.where(df['value_type'] == "humidity", df["value"], 'NaN')
    df['height'] = np.where(df['value_type'] == 'height', df["value"], 'NaN')
    
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    
    convert_dict = {'value': float,'P2': float,'P1':float,'P0':float,'temperature':float,'humidity':float,'height':float}
    df = df.astype(convert_dict)
    
    df = df.drop(columns=['value'], axis=1)
    
    df = df.rename(columns={"sensor_id": "metadata.sensor_id",
                            "sensor_type": "metadata.sensor_type",
                            "location":"metadata.location",
                            "lat":"metadata.lat",
                            "lon":"metadata.lon",
                            "value_type":"metadata.measrment"})
    return df
```

```python
frame = wrangle(r"E:\WorldQuant\project_3\data\november_2019_sensor_data_archive.csv")
```
